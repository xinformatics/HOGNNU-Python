## Graph Measures

### Centrality

#### Degree centrality = simply defined as degree of the node
#### Closeness centrality = avg length of shorttest path to this node from all other nodes
#### Betweenness centrality = of all the shortest paths between 
#### pairs of node, how many have this particular node in them.

## Some graph traversal algorithms BFS and DFS

```python
## we study these algorithms on our graph G

## lets print G once

nx.draw(G, with_labels=True, font_color='#ffffff', pos=pos_G)
```

```python
### to start  bfs we need the graph on which we run bfs and the starting node

def bfs(graph, node):
    
    visited, queue = [node], [node]
    
    while queue:
        node = queue.pop(0)
        
        for neighbour in graph[node]: 
            ##  graph[node] returns the connectivity information
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)
                
    return visited
```

```python
bfs(G, 'A')
```

```python
bfs(G, 'D')
```

### BFS is useful:
### 1: shorttest path between two nodes in unweighted graph
### 2: can be used to check graph connectivity / find connected components
### 3: time complexity is O(|V| + |E|)

### BFS variants
#### bidirectional BFS
#### A*

## DFS - Depth First Search
### recursive algorithm

```python
### print our graph again
nx.draw(G, with_labels=True, font_color='#ffffff', pos=pos_G)
```

```python
visited = []

def dfs(visited, graph, node):
    if node not in visited:
        visited.append(node)
        
        for neighbour in graph[node]:
            visited = dfs(visited, graph, neighbour)
            
    return visited
```

```python
dfs([], G, 'D')
```

```python
dfs([], G, 'A')
```

### DFS is useful:
### 1: Finding connected components
### 2: Finding graph cycles
### 3: Topological Sorting

#### time complexity O(|V| + |E|)

#### can get trapped in infinite loops in DFS
